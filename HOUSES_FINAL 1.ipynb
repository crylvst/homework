```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("data/train.csv")
df_test = pd.read_csv("data/test.csv")
```

```python
# Дропаем ненужное
df_train = df_train.drop(columns=['PoolQC','MiscFeature', 'Alley', 'Fence', 'PoolArea', 'GarageYrBlt', 'YearRemodAdd', 'Utilities'], axis=1)
df_test = df_test.drop(columns=['PoolQC','MiscFeature', 'Alley', 'Fence', 'PoolArea', 'GarageYrBlt', 'YearRemodAdd', 'Utilities'], axis=1)
```

```python
# разделяем выборку на два датафрейма по типу данных
df_train_obj=pd.DataFrame()
df_train_int=pd.DataFrame()
l=[]
r=[]
for i in df_train:
  if(np.dtype(df_train[i])=='object'):
    l.append(i)
    df_train_obj[i]=df_train[i]
  else:
    df_train_int[i]=df_train[i]
    if(np.dtype(df_train[i])=='int64'):
      r.append(i)

df_test_obj=pd.DataFrame()
df_test_int=pd.DataFrame()
l=[]
r=[]
for i in df_test:
  if(np.dtype(df_test[i])=='object'):
    l.append(i)
    df_test_obj[i]=df_test[i]
  else:
    df_test_int[i]=df_test[i]
    if(np.dtype(df_test[i])=='int64'):
      r.append(i)
```

```python
# импортируем словарь с данными, где признакам объекта присваивается число
import ast
import json
file = open("data/dict_train.txt", "r")

contents = file.read()
dict_train = json.loads(contents)
```

```python
# заменяем во всем датафрейме объекты на числа из словаря
for column in dict_train:
    df_train_obj[column] = df_train_obj[column].map(dict_train[column])
```

```python
for column in dict_train:
    df_test_obj[column] = df_test_obj[column].map(dict_train[column])
```

```python
# заменяем пустоту на ноль
df_train_int.fillna(0, inplace=True)
df_train_obj.fillna(0, inplace=True)
df_test_int.fillna(0, inplace=True)
df_test_obj.fillna(0, inplace=True)
```

```python
df_train_int = df_train_int.astype('int64')
df_train_obj = df_train_obj.astype('int64')
```

```python
# объединяем дфы
df_for_ml = df_train_int.join(df_train_obj)
df_for_test = df_test_int.join(df_test_obj)
```

```python
x = df_for_ml.drop(['SalePrice'], axis=1)
y = np.log(df_train['SalePrice'])
z = df_for_test
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(x, y)
predicted_svr = svr.predict(z)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x,y)
predicted_dtr = dtr.predict(z)
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(x, y)
predicted_xgb = xgb.predict(z)
```

```python
from catboost import CatBoostRegressor
catb = CatBoostRegressor()
catb.fit(x, y)
predicted_catb = catb.predict(z)
```

```python
# стекинг моделей
from sklearn.ensemble import StackingRegressor

estimators = [
    ("svr", svr),
    ("dtr", dtr),
    ("xgb", xgb),
    ("catb", catb)
]

sr = StackingRegressor(estimators=estimators)

sr.fit(x, y)
predicted_sr = sr.predict(z)
```

```python
# создаем датафрейм, куда заливаем предсказанные цены
result=pd.DataFrame()
result['Id'] = range(1461,2920)
```

```python
# загружаем цены в датафрейм
final_sr=pd.DataFrame()
final_sr['SalePrice'] = predicted_sr
result=result.join(final_sr)
```

```python
# применяем антилогарифм, потому что цена предсказывалась в бустере в логарифме
result['SalePrice'] = np.exp(result['SalePrice'])
```

```python
# сохраняем в csv
result.to_csv("houses final.csv",index=False)
```
